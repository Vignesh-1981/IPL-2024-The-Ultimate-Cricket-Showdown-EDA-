```python
# Step 0: Clean column names (strip whitespace)
data.columns = data.columns.str.strip()

# Step 1: Check which of your expected columns are missing
numeric_cols = ['Year', 'Matches_Batted', 'Not_Outs', 'Runs_Scored', 'Highest_Score',
                'Batting_Average', 'Balls_Faced', 'Batting_Strike_Rate', 'Centuries',
                'Half_Centuries', 'Fours', 'Sixes', 'Catches_Taken', 'Stumpings',
                'Matches_Bowled', 'Balls_Bowled', 'Runs_Conceded', 'Wickets_Taken',
                'Best_Bowling_Match', 'Bowling_Average', 'Economy_Rate',
                'Bowling_Strike_Rate', 'Four_Wicket_Hauls', 'Five_Wicket_Hauls']

# Step 2: Identify missing columns
missing_cols = [col for col in numeric_cols if col not in data.columns]
print("Missing columns:", missing_cols)

# Step 3: Apply conversion only to existing columns
existing_numeric_cols = [col for col in numeric_cols if col in data.columns]
data[existing_numeric_cols] = data[existing_numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load your data
data = pd.read_csv('/kaggle/input/ipl-dataset-upto-2022/playerlifetime/cricket_data.csv')

# Step 2: Clean column names by stripping leading/trailing whitespace
data.columns = data.columns.str.strip()

# Optional: Confirm 'Runs_Scored' exists
if 'Runs_Scored' not in data.columns:
    print("Column 'Runs_Scored' not found! Available columns are:")
    print(data.columns.tolist())
else:
    # Step 3: Convert Runs_Scored to numeric (in case it's object type)
    data['Runs_Scored'] = pd.to_numeric(data['Runs_Scored'], errors='coerce')

    # Step 4: Plot the histogram
    plt.figure(figsize=(12, 8))
    sns.histplot(data=data, x='Runs_Scored', bins=20, kde=True)
    plt.title('Distribution of Runs Scored')
    plt.xlabel('Runs Scored')
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Step 5: Further Analysis
top_run_scorers = data.sort_values(by='Runs_Scored', ascending=False).head(10)
print("Top Run Scorers:\n", top_run_scorers)

top_wicket_takers = data.sort_values(by='Wickets_Taken', ascending=False).head(10)
print("Top Wicket Takers:\n", top_wicket_takers)

plt.figure(figsize=(12, 8))
sns.lineplot(data=data, x='Year', y='Runs_Scored')
plt.title('Trend of Runs Scored Over Years')
plt.xlabel('Year')
plt.ylabel('Runs Scored')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Exclude non-numeric columns before computing correlation
numeric_data = data.select_dtypes(include=['int', 'float'])

# Compute correlation matrix
correlation_matrix = numeric_data.corr()

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of IPL Data')
plt.show()
```

```python
# Visualize Runs Scored distribution
plt.figure(figsize=(10, 6))
plt.hist(data['Runs_Scored'], bins=20, color='orange', edgecolor='black')
plt.xlabel('Runs Scored')
plt.ylabel('Frequency')
plt.title('Distribution of Runs Scored')
plt.show()
```

```python
# Sort the DataFrame by 'Runs_Scored' column in descending order
data_sorted = data.sort_values(by='Runs_Scored', ascending=False)

# Select the top ten players with the highest runs scored
top_ten_players = data_sorted.head(10)

# Visualize Runs Scored by the top ten players
plt.figure(figsize=(12, 6))
plt.bar(top_ten_players['Player_Name'], top_ten_players['Runs_Scored'], color='skyblue')
plt.xlabel('Player Name')
plt.ylabel('Runs Scored')
plt.title('Top Ten Players by Runs Scored')
plt.xticks(rotation=45, ha='right') 
plt.show()
```
